# 4 , Perform dimensionality reduction.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Separate the target variable (if applicable) from the features:
X = train_df.drop(['y'], axis = 1)
y = train_df['y']
```

```python
# Standardize the feature variables:
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Apply PCA to reduce the dimensionality of the data:
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)
```

```python
X_pca.shape
```

```python
type(X_pca)
```

# 5,Predict your test_df values using xgboost

```python
# Xg boost model building
from sklearn.model_selection import train_test_split
X_train, X_test , y_train, y_test = train_test_split(X_pca,y , test_size=0.25 , random_state=77)
```

```python
import scipy
```

```python
# Instantiation
import xgboost as xg
xgb_r = xg.XGBRegressor(objective ='reg:squarederror',n_estimators = 10, seed = 123)
```

```python
xgb_r.fit(X_train,y_train)
y_pred = xgb_r.predict(X_test)
```

```python
d = pd.DataFrame()
d["test_y"] = y_test
d["y_pred"] = y_pred
d["mp"] = abs((d["test_y"]- d["y_pred"])/d["test_y"])
(d.mp.mean())*100#mape
print(100-(d.mp.mean())*100)
```
